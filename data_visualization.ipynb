```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
```

```python
train['actual'].isna().sum()
```

```python
test['actual'].isna().sum()
```

```python
train_moving = train.where(train == 'moving')['actual'].count()
train_stationary = train.where(train == 'stationary')['actual'].count()

train_moving, train_stationary, train_stationary+train_moving, train.shape[0]
```

```python
plt.pie([train_moving, train_stationary], labels=['Moving', 'Stationary'], startangle=90, autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Label distribution')
plt.plot()
```

```python
test_moving = test.where(test == 'moving')['actual'].count()
test_stationary = test.where(test == 'stationary')['actual'].count()
```

```python
plt.pie([test_moving, test_stationary], labels=['Moving', 'Stationary'], startangle=90, autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Label distribution')
plt.plot()
```
